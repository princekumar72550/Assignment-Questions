```python
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
```

```python
Scatter Plot:
Use: Display the relationship between two numeric variables.
Function: seaborn.scatterplot()
Example: Useful for identifying patterns, trends, and outliers in bivariate data.

Line Plot:
Use: Display the trend of a numeric variable over a continuous interval or time.
Function: seaborn.lineplot()
Example: Suitable for visualizing time series data or any data where the x-axis represents a continuous variable.

Bar Plot:
Use: Compare the values of categorical variables using rectangular bars.
Function: seaborn.barplot()
Example: Useful for comparing the mean or aggregate values of different categories.

Histogram:
Use: Display the distribution of a single variable.
Function: seaborn.histplot()
Example: Helps in understanding the underlying frequency distribution of a continuous variable.

Box Plot (Box-and-Whisker Plot):
Use: Summarize the distribution of a continuous variable and identify potential outliers.
Function: seaborn.boxplot()
Example: Useful for comparing the spread and central tendency of different categories.
```

```python
Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = 
"timepoint" and y = "signal" for different events and regions.Â 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fmri_data = sns.load_dataset("fmri")

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)

plt.title("FMRI Signal Over Time for Different Events and Regions")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.show()
```

```python
Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 
'pclass', y = 'age' and y = 'fare'.

Note:Â  pclass, age, and fare are columns in the titanic dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset("titanic")

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

sns.boxplot(x='pclass', y='age', data=titanic_data, ax=axes[0])
axes[0].set_title('Box Plot of Age by Pclass')

sns.boxplot(x='pclass', y='fare', data=titanic_data, ax=axes[1])
axes[1].set_title('Box Plot of Fare by Pclass')

plt.tight_layout()

plt.show()
```

```python
Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue 
parameter for the 'cut' column of the diamonds dataset.Â 
```

```python
import seaborn as sns
diamonds = sns.load_dataset('diamonds')
sns.histplot(data=diamonds, x='price', hue='cut')
plt.title("Histogram of Diamond Prices with Cut Differentiation")
plt.xlabel("Price")
plt.ylabel("Frequency")

plt.show()
```
